## Communication is Key

Although many many students choose to study physics due to the perception it is heavily based on experimentation and mathematics, one of the key skills a physicist has is their ability to communicate their research. A research academic in our department must be able to communicate with various shareholders. **Consider who these groups are and how they differ, then take a look at the dropdown.** 

```{admonition} Who are the different shareholders a physicist must be able to communicate with?
:class: dropdown

**Their Colaboration:** Physicists tend to work in research clusters and larger collaborations. This is because we are often working towards one common goal. It is therefore very important to be able to communicate with your collegues about the work you are doing and why in order to make a meaningful impact. Ineffective communication could leave your peers confused or uninformed about what you are working on, which may ultimately slow down progress for the enitre group e.g. if I do not understand your work, I may repeat tasks unecessarily or work on incorrect assumptions.



**The Research Community:** In order for your contribution to the field to be recognised and for incremental progress in the field to be made, it is key that you are able to share your work with the wider research community. Whereas those in your cluster may be very familiar with your work, others in the wider field may not be. As such, a competent scientist must be able to clearly articulate what they have achieved and how so that others are able to understand and build upon the work.



**Their Students:** Not only does an academic perform research, but they may also teach students in undergraduate courses or have their own PhD students. When sharing ideas with these groups, the academic will need to be able to pitch at the appropriate level. Although both a fourth year PhD student and a first year undergraduate student will have an interest in physics, they will posses very different levels of knowledge and experience. As such it would not be appropriate to assume that an both would understand the same explaination - the academic must make considerations in how they commuinicate with each. 



**The Media:** Large scale physics experiements tend to be very costly, which introduces a social contract between scientists and the public, after all, much of the funding for the research comes from taxation. It is therefore key that we are can explain to the public what the work we are doing is and why it is important. Unlike in the examples above, the audience is no longer a group of trained scientists interested in your work, again it is important that the information is pitched to the correct level. This is a skill in itself.
```

The list of shareholders above is not exhaustive, but provides an overview of why sharing our research appropriately and effectively is important. Depending on who we are deceminating information to will also dictate which format we do so in. **Consider the formats that a scientist may share their work, then take a look at the dropdown.**

```{admonition} What are the formats that a scientist may share their work?
:class: dropdown

**General Audience**
- News article 
- Interview 
- Popular science publications (New Scientist etc.)
- Podcast
- Documentary
- Outreach events (e.g. Pint of Science!)

**Scientific Audience**

- Lab report
- Disertation
- PhD thesis
- Conference talk
- Scientific poster
- Publication (e.g. Nature, Physics Review Letters etc.)

```

During your undergraduate course you will give conference style and poster presentations, however, you will gain the most experience in report writing. However, before we begin learning how to write and strucutre good reports, it is worth learning how to source information. This will also give you the opportunity to identify the differences between sources of information.

### Pear Shaped Nuclei (30 mins)

Let us begin by taking at the difference between the different formats that scientists employ for sharing their research. In 2013, an international team researching the structures of nuclei uncovered evidence for pear-shaped nuclei. Although it has been established for decades that nuclei are often symetrically deformed, this work provided evidence for *octupole* deformed nuclei, which has profound implications for our understanding of the universe, including the possibility of time travel. This work was reported in many different formats in various outlets.

Review each of these articles on the topic. For each identify who the target audience is and note down the major differences between each format.
